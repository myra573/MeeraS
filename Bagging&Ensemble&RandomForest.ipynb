```python
# 1. Create the Dataset
data = {
    'ID': [1, 2, 3, 4, 5],
    'Weight': [20, 35, 5, 50, 2],
    'Texture': ['Fluffy', 'Smooth', 'Fluffy', 'Smooth', 'Smooth'],
    'Label': ['Yes', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)
```

```python
def create_bootstrap_sample(dataset):
    n_samples = len(dataset)
    indices = []

    for i in range(n_samples):
      random_index = random.randint(0, n_samples - 1)
      indices.append(random_index)

    bootstrap_sample = dataset.iloc[indices]
    return bootstrap_sample
```

```python
def majority_vote(predictions):
  vote_counts = {}

  for prediction in predictions:
    if prediction in vote_counts:
      vote_counts[prediction] += 1

    else:
      vote_counts[prediction] = 1

  print(vote_counts)
```

```python
# Comparing accuracy of single DT vs Bagging Model on real world Dataset -> Wine_Dataset
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
```

```python
data = load_wine()
X = data.data
y = data.target
```

```python
data = pd.DataFrame(X, columns=data.feature_names)
data['target'] = y
data
```

```python
# Lists to store accuracies for every run
tree_accuracies = []
bagging_accuracies = []

# 2. Run the experiment 10 times
print("Run | Tree Acc | Bagging Acc")
print("----|----------|------------")

for i in range(10):
    # vary random_state by 'i' to get different splits each time
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # --- Single Decision Tree ---
    # We use a deep tree (max_depth=None) to ensure it has high variance potential
    tree = DecisionTreeClassifier(max_depth=None, min_samples_split=2)
    tree.fit(X_train, y_train)
    tree_acc = accuracy_score(y_test, tree.predict(X_test))
    tree_accuracies.append(tree_acc)

    # --- Bagging Classifier ---
    # An ensemble of 20 such trees
    bag = BaggingClassifier(estimator=tree, n_estimators=20, random_state=42)
    bag.fit(X_train, y_train)
    bag_acc = accuracy_score(y_test, bag.predict(X_test))
    bagging_accuracies.append(bag_acc)

    print(f" {i+1:2d} |  {tree_acc:.4f}  |   {bag_acc:.4f}")
```

```python
# OOB -> Out of Bag ->Samples which never got selected in random selection

# Single DT ->  low bias / high variance (overfitting)
# Bagging ->    high bias / low variance (can lead to underfitting)
# Explanation
```

# Random Forest

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=10, random_state=42, max_features='sqrt', oob_score=True)
```

```python
model.fit(X,y)
```

```python
print("train_accuracy:", accuracy_score(y_train, model.predict(X_train)))
print("test_accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
print("oob_score:", model.oob_score_)
```

```python
# Why it is not overfitting?
# Ensemble (combination of various trees) and each tree is looking at a different feature to make prediction
```

```python
# Feature Importance
importances = model.feature_importances_
```

```python
df.columns
```

```python
indices = np.argsort(importances)[::-1] # Sort descending
# print("\nTop 5 Important Features:")
for i in range(5):
    print(f"{i+1}. {df.columns[indices[i]]}: {importances[indices[i]]:.4f}")
```

```python
# Challenge : Apply Random forest on Titanic Dataset and Reveal feature importance
# Which was the most important feature for survival?
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df = df[['survived', 'pclass', 'sex', 'age', 'fare']]
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df = pd.get_dummies(df, columns=['sex'])
```

```python
df
```

```python
X = df[['pclass', 'sex_male','sex_female', 'age', 'fare']]
y = df['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42, max_features='sqrt')
```

```python
model.fit(X,y)
```

```python
importances = model.feature_importances_
```

```python
indices = np.argsort(importances)[::-1] # Sort descending
# print("\nTop 5 Important Features:")
for i in range(5):
    print(f"{i+1}. {df.columns[indices[i]]}: {importances[indices[i]]:.4f}")
```
