```python
# Q6 Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Display the regression equation
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

print(f"Regression Equation: Y = {model.intercept_:.2f} + {model.coef_[0]:.2f}X")

# Plot the data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')

# Add labels and title
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Display the plot
plt.show()
```

```python
# Q7 # Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create the dataset
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent variables
X = df[['Area', 'Rooms']]

# Dependent variable
y = df['Price']

# Train the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
predictions = model.predict(X)

# Display coefficients
print("Intercept:", model.intercept_)
print("Coefficients:")
print("Area :", model.coef_[0])
print("Rooms:", model.coef_[1])

# Regression Equation
print("\nRegression Equation:")
print(f"Price = {model.intercept_:.2f} + ({model.coef_[0]:.2f} × Area) + ({model.coef_[1]:.2f} × Rooms)")

# Calculate Variance Inflation Factor (VIF)
vif = pd.DataFrame()
vif["Feature"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif)
```

```python
# Q8 Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train the Polynomial Regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
Y_pred = model.predict(X_poly)

# Display coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Generate smooth curve for plotting
X_new = np.linspace(1, 5, 100).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
Y_new = model.predict(X_new_poly)

# Plot original data
plt.scatter(X, Y, color='blue', label='Actual Data')

# Plot polynomial curve
plt.plot(X_new, Y_new, color='red', linewidth=2, label='Polynomial Regression Curve')

# Labels and title
plt.title("2nd Degree Polynomial Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Display plot
plt.show()
```

```python
# Q9 # Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train the Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Print regression equation
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])

# Print residuals
print("\nResiduals:")
for actual, predicted, residual in zip(Y, Y_pred, residuals):
    print(f"Actual: {actual}, Predicted: {predicted:.2f}, Residual: {residual:.2f}")

# Create Residual Plot
plt.figure(figsize=(6,5))
plt.scatter(Y_pred, residuals)

# Horizontal line at zero
plt.axhline(y=0, color='red', linestyle='--')

plt.title("Residual Plot")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")

plt.show()
```
